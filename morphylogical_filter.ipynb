```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

source = cv2.imread('data/source.png', cv2.IMREAD_GRAYSCALE)
destination = cv2.imread('data/destination.png', cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Source Image")
plt.imshow(source, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Destination Image")
plt.imshow(destination, cmap='gray')
plt.axis('off')

plt.show()
```

```python
# OPEN ( A open B = (A erode B) dilate B )
# CLOSE ( A close B = (A dilate B) erode B )
```

```python
kernel1 = np.array([[1, 1, 0],
                   [1, 1, 0],
                   [0, 0, 0]], dtype=np.uint8)


dilate_image = cv2.dilate(source, kernel1, iterations=1)

kernel2 = np.array([[0, 1, 1],
                   [0, 1, 1],
                   [0, 0, 0]], dtype=np.uint8)

done_image = cv2.erode(dilate_image, kernel2, iterations=1)


plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Source Image")
plt.imshow(source, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Transformed Image")
plt.imshow(done_image, cmap='gray')
plt.axis('off')
```
